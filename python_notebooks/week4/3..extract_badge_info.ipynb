```python
import pandas as pd
import ast
```

```python
# read df_master_clean.csv from processed_data folder
input_file_path = "../../data/processed_silver/week4/esl_dota2_dream_league_26_emote_panel_v1.csv"
df = pd.read_csv(input_file_path)
df
```

```python
# convert 'i_badges' column of type string to python list object
df["badge_info"] = df["i_badges"].apply(ast.literal_eval)
```

```python
# define function to extract "i_badge_names", "i_badge_titles", "i_badge_versions" from "i_badges" column 
def extract_badge_info(badge_list):
    if not isinstance(badge_list, list):
        return "","",""
    names = []
    titles = []
    versions = []
    for badge in badge_list:
        names.append(badge.get("name", ""))
        titles.append(badge.get("title", ""))
        versions.append(str(badge.get("version","")))
    return ", ".join(names),", ".join(titles), ", ".join(versions)
```

```python
new_columns = ["i_badge_names", "i_badge_titles", "i_badge_versions"]
```

```python
df[new_columns] = df["badge_info"].apply(
    lambda x: pd.Series(extract_badge_info(x)))
```
